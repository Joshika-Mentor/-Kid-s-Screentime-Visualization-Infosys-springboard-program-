```python
df['Gender']=df['Gender'].astype('category')
df['Urban_or_Rural']=df['Urban_or_Rural'].astype('category')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], kde=True, color='skyblue')
plt.title('distribution of screen time in hours')
plt.show()
```

```python
# Violin plot of Screen Time by Gender
plt.figure(figsize=(8,5))
sns.violinplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df, palette='Set2')
plt.title('Average Daily Screen Time by Gender')
plt.xlabel('Gender')
plt.ylabel('Avg Daily Screen Time (hr)')
plt.show()
```

```python
# Screen time distribution by Primary Device
plt.figure(figsize=(8,5))
sns.violinplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df, palette='cool')
plt.title('Average Daily Screen Time by Age')
plt.xlabel('Age')
plt.ylabel('Avg Daily Screen Time (hr)')
plt.show()
```

```python
# Screen time by Urban or Rural background
plt.figure(figsize=(8,5))
sns.violinplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', data=df, palette='Set3')
plt.title('Screen Time Distribution by Region Type')
plt.xlabel('Urban or Rural')
plt.ylabel('Avg Daily Screen Time (hr)')
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[10, 20, 30, 40, 50, 60, 70],
                        labels=['1020', '2130', '3140', '4150', '5160', '6170'])
```

```python
plt.figure(figsize=(14, 6))

# Boxplot: Screen Time by Age Group
plt.subplot(1, 2, 1)
sns.boxplot(x='AgeGroup', y='Avg_Daily_Screen_Time_hr', data=df, palette='coolwarm')
plt.title(' Screen Time by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Avg Daily Screen Time (hrs)')
```

```python
# Boxplot: Screen Time by Gender
plt.subplot(1, 2, 2)
sns.boxplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df, palette='Set2')
plt.title(' Screen Time by Gender')
plt.xlabel('Gender')
plt.ylabel('Avg Daily Screen Time (hrs)')

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

# Histogram and KDE for Screen Time
plt.figure(figsize=(8,5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], kde=True, color='skyblue', bins=15)
plt.title('Distribution of Average Daily Screen Time')
plt.xlabel('Average Daily Screen Time (hrs)')
plt.ylabel('Number of Kids')
plt.show()
```

```python
# Histogram and KDE for Age
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], kde=True, color='lightgreen', bins=10)
plt.title('Distribution of Age')
plt.xlabel('Age (years)')
plt.ylabel('Number of Kids')
plt.show()
```

```python
# Scatter plot of Age vs Average Daily Screen Time
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df, color='teal', s=70, edgecolor='black')

plt.title('Scatter Plot: Age vs Average Daily Screen Time')
plt.xlabel('Age (years)')
plt.ylabel('Average Daily Screen Time (hrs)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np
import pandas as pd
dummy_dates = pd.to_datetime(pd.date_range(start='2024-01-01', periods=len(df),freq='D'))
df['Date'] = dummy_dates
df['Day_type'] = np.where(df['Date'].dt.dayofweek>=5, 'Weekday', 'Weekend')
display(df.head())
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Gender', y='Avg_Daily_Screen_Time_hr', hue='Day_type', palette='pastel')
plt.title('Average Daily Screen Time by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Daily Screen Time (hours)')
plt.show()
```

```python
g = sns.FacetGrid(df, col='Primary_Device', row='Gender', height=4, aspect=1.2)
g.map_dataframe(sns.histplot, x='Avg_Daily_Screen_Time_hr', color='skyblue', kde=True, bins=10)
g.set_axis_labels("Average Daily Screen Time (hrs)", "Count")
g.set_titles(col_template="{col_name}", row_template="{row_name}")
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Distribution of Screen Time by Device and Gender')
plt.show()
```

```python
# Selecting only numeric columns
numeric_cols = ['Age', 'Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']

# Pairplot to show relationships
sns.pairplot(df[numeric_cols], diag_kind='kde', palette='coolwarm')
plt.suptitle('Pairplot: Relationships Between Numerical Variables', y=1.02)
plt.show()
```

```python
numeric_cols = ['Age', 'Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']
# correlation matrix
corr = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='Avg_Daily_Screen_Time_hr', hue='Gender',
             kde=True, bins=30, palette='Set2', alpha=0.6)
plt.title('Distribution of Daily Screen Time by Gender')
plt.xlabel('Average Daily Screen Time (hours)')
plt.ylabel('Count')
plt.show()
```

```python
device_location = df.pivot_table(index='Urban_or_Rural',
                                 columns='Primary_Device',
                                 aggfunc='size',
                                 fill_value=0)

# Plot stacked bar chart
device_location.plot(kind='bar', stacked=True, figsize=(8,6), colormap='tab20')
plt.title('Device Usage by Location')
plt.xlabel('Location')
plt.ylabel('Number of Users')
plt.legend(title='Primary Device', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr',
              hue='Gender', palette='pastel', size=2, dodge=True)
plt.title('Screen Time Spread by Location and Gender')
plt.xlabel('Location')
plt.ylabel('Average Daily Screen Time (hours)')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(data=df, x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr',
              hue='Gender', palette='Set2', jitter=0.25, dodge=True)
plt.title('Screen Time Spread by Location and Gender')
plt.xlabel('Location')
plt.ylabel('Average Daily Screen Time (hours)')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1))
plt.show()
```
