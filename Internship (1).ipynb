```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set(style="whitegrid")

# --- Clean column names ---
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
print("Cleaned column names:", df.columns.tolist())
# Output: ['age', 'gender', 'avg_daily_screen_time_hr', ...]

# --- Ensure screen time column is numeric ---
df['avg_daily_screen_time_hr'] = pd.to_numeric(df['avg_daily_screen_time_hr'], errors='coerce')

# Drop NaN values for plotting
df_plot = df.dropna(subset=['avg_daily_screen_time_hr'])

# --- Violin Plot: Screen Time by Gender ---
plt.figure(figsize=(7,5))
sns.violinplot(data=df_plot, x='gender', y='avg_daily_screen_time_hr', palette='pastel')
plt.title('Average Daily Screen Time by Gender')
plt.xlabel('Gender')
plt.ylabel('Screen Time (hours)')
plt.tight_layout()
plt.show()
# --- Violin Plot: Screen Time by Age ---
plt.figure(figsize=(7,5))
sns.violinplot(data=df_plot, x='age', y='avg_daily_screen_time_hr', palette='muted')
plt.title('Average Daily Screen Time by Age')
plt.xlabel('Age')
plt.ylabel('Screen Time (hours)')
plt.tight_layout()
plt.show()
```

**Correlation Heatmap**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load CSV (ensure file is uploaded or correct path)
df = pd.read_csv("/content/archive.zip")

# Step 2: Clean column names
df.columns = df.columns.str.strip().str.lower()

# Step 3: Select only numerical columns for correlation
numerical_cols = ['age', 'avg_daily_screen_time_hr', 'educational_to_recreational_ratio']

# Step 4: Compute correlation matrix
corr = df[numerical_cols].corr()

# Step 5: Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap: Numerical Variables")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
np.random.seed(0)
n = 500

df = pd.DataFrame({
    'screen_time': np.concatenate([
        np.random.normal(6, 1.5, size=n),  # Simulated screen time for males
        np.random.normal(5, 1.2, size=n)   # Simulated screen time for females
    ]),
    'gender': ['Male'] * n + ['Female'] * n
})

# Plot
sns.histplot(data=df, x='screen_time', hue='gender', kde=True, element='step', stat='density', common_norm=False)
plt.title('Distribution of Screen Time by Gender')
plt.xlabel('Screen Time (hours)')
plt.ylabel('Density')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Location': ['NY', 'LA', 'Chicago'],
    'Mobile': [100, 80, 90],
    'Tablet': [40, 30, 50],
    'Desktop': [60, 50, 70]
}

df = pd.DataFrame(data)
df.set_index('Location', inplace=True)

# Plot stacked bar chart
ax = df.plot(kind='bar', stacked=True, figsize=(8,6))

plt.title('Device Usage by Location')
plt.xlabel('Location')
plt.ylabel('Number of Users')
plt.legend(title='Device')
plt.tight_layout()
plt.show()
```

**Swarmplots/stripplots(Screen time sprea**d)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'User': range(1, 21),
    'ScreenTime': [3.5, 5.0, 2.0, 4.5, 6.0, 2.5, 3.0, 4.0, 5.5, 3.7,
                   2.8, 4.2, 5.1, 3.3, 6.5, 2.2, 3.8, 4.7, 5.3, 3.1],
    'Device': ['Mobile', 'Desktop', 'Tablet', 'Mobile', 'Desktop',
               'Tablet', 'Mobile', 'Desktop', 'Tablet', 'Mobile',
               'Desktop', 'Tablet', 'Mobile', 'Desktop', 'Tablet',
               'Mobile', 'Desktop', 'Tablet', 'Mobile', 'Desktop']
}

df = pd.DataFrame(data)

# Swarmplot
plt.figure(figsize=(8,5))
sns.swarmplot(x='Device', y='ScreenTime', data=df)
plt.title('Screen Time Spread by Device (Swarmplot)')
plt.show()

# Stripplot (alternative)
plt.figure(figsize=(8,5))
sns.stripplot(x='Device', y='ScreenTime', data=df, jitter=True)
plt.title('Screen Time Spread by Device (Stripplot)')
plt.show()
```

**Combine multiple plots into subplot**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'User': range(1, 21),
    'ScreenTime': [3.5, 5.0, 2.0, 4.5, 6.0, 2.5, 3.0, 4.0, 5.5, 3.7,
                   2.8, 4.2, 5.1, 3.3, 6.5, 2.2, 3.8, 4.7, 5.3, 3.1],
    'Device': ['Mobile', 'Desktop', 'Tablet', 'Mobile', 'Desktop',
               'Tablet', 'Mobile', 'Desktop', 'Tablet', 'Mobile',
               'Desktop', 'Tablet', 'Mobile', 'Desktop', 'Tablet',
               'Mobile', 'Desktop', 'Tablet', 'Mobile', 'Desktop']
}

df = pd.DataFrame(data)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Swarmplot on first subplot
sns.swarmplot(x='Device', y='ScreenTime', data=df, ax=axes[0])
axes[0].set_title('Screen Time Spread by Device (Swarmplot)')

# Stripplot on second subplot
sns.stripplot(x='Device', y='ScreenTime', data=df, jitter=True, ax=axes[1])
axes[1].set_title('Screen Time Spread by Device (Stripplot)')

plt.tight_layout()
plt.show()
```
