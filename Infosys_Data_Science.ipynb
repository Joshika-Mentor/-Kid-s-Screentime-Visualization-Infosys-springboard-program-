```python
import pandas as pd
df="Indian_Kids_Screen_Time.csv"
df=pd.read_csv(df)
df.head()
#df.tail()
#df.tail(10) Last 10 observations
```

```python
df.columns
```

```python
df.describe()
```

```python
df.shape #rows and columns
```

```python
df=df["Primary_Device"] #particular column
df
```

```python
df=df[["Primary_Device","Health_Impacts"]]
df
```

```python
df=df[2:6:1]
df
```

```python
df=df[5:50:5]
df
```

```python
print(df.isnull().sum())
```

```python
df['Gender']=df['Gender'].astype('category')
df['Urban_or_Rural']=df['Urban_or_Rural'].astype('category')
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'],kde=True,color="skyblue")
plt.title("Distribution of screen time in hours")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 4))

# Histogram of Average Daily Screen Time
plt.subplot(1, 2, 1)
sns.histplot(df['Avg_Daily_Screen_Time_hr'], kde=False, color='red')
plt.title('Histogram of Average Daily Screen Time')
plt.xlabel('Average Daily Screen Time ')
plt.ylabel('Frequency')

# Distribution of Age
plt.subplot(1, 2, 2)
sns.histplot(df['Age'], kde=False, color='yellow')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 4))
sns.countplot(x='Gender', data=df)
plt.title('Count of Users by Gender')
plt.show()

sns.countplot(x='Primary_Device', data=df)
plt.title('Count of Users by Device Type')
plt.show()

sns.countplot(x='Urban_or_Rural', data=df)
plt.title('User Distribution by Location')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 5))
sns.boxplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title('Screen Time by Gender (Boxplot)')
plt.show()

sns.violinplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title('Screen Time by Gender (Violin Plot)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 5))

# KDE Plot of Average Daily Screen Time
plt.subplot(1, 2, 1)
sns.kdeplot(df['Avg_Daily_Screen_Time_hr'], fill=True, color='skyblue')
plt.title('KDE Plot of Average Daily Screen Time')
plt.xlabel('Average Daily Screen Time (hrs)')

# KDE Plot of Age
plt.subplot(1, 2, 2)
sns.kdeplot(df['Age'], fill=True, color='lightgreen')
plt.title('KDE Plot of Age')
plt.xlabel('Age')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df, color='coral')

plt.title('Scatter Plot of Age vs Average Daily Screen Time')
plt.xlabel('Age')
plt.ylabel('Average Daily Screen Time (hrs)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
sns.barplot(x='Gender', y='Avg_Daily_Screen_Time_hr', hue='Primary_Device', data=df)

plt.title('Grouped Bar Chart: Screen Time by Gender and Device Type')
plt.xlabel('Gender')
plt.ylabel('Average Daily Screen Time (hrs)')
plt.legend(title='Device Type')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.histplot(data=df, x='Avg_Daily_Screen_Time_hr',hue='Gender',kde=True, bins=30)

plt.title('Distribution of Screen Time by Gender')
plt.xlabel('Average Daily Screen Time (hrs)')
plt.ylabel('Frequency')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
g = sns.FacetGrid(df, col='Primary_Device', hue='Gender', height=4)
g.map(sns.histplot, 'Avg_Daily_Screen_Time_hr', kde=True).add_legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.pairplot(df[['Age', 'Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
corr = df[['Age','Avg_Daily_Screen_Time_hr','Educational_to_Recreational_Ratio']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
pd.crosstab(df['Urban_or_Rural'], df['Primary_Device']).plot(kind='bar', stacked=True)
plt.title('Stacked Barplot: Device Usage by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.stripplot(
    x='Gender',
    y='Avg_Daily_Screen_Time_hr',
    data=df,
    jitter=True,
    alpha=0.7
)
plt.title('Screen Time Spread by Gender')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')

plt.subplot(1,2,2)
sns.histplot(df['Avg_Daily_Screen_Time_hr'], kde=True)
plt.title('Screen Time Distribution')

plt.tight_layout()
plt.show()
```
