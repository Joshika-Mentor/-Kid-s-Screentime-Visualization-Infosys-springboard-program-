# Day 2: FacetGrid(Screen time by device X gender)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the FacetGrid
g = sns.FacetGrid(df, col="Gender", height=5, aspect=1)

# Map the barplot to the grid (assign hue same as x)
g.map_dataframe(
    sns.barplot,
    x="Primary_Device",
    y="Avg_Daily_Screen_Time_hr",
    hue="Primary_Device",      # assign hue same as x
    palette="Set1",
    legend=False
)

# Set axis labels
g.set_axis_labels("Primary Device", "Average Daily Screen Time (hr)")

# Set titles for each facet
g.set_titles(col_template="{col_name}")

# Add a main title
g.fig.suptitle("Screen Time by Device × Gender", fontsize=14, y=1.05)


# Show the plot
plt.show()
```

Insights: The visualization demonstrates that average daily screen time for children is highly consistent across all primary devices and between both genders, consistently remaining around 4.3 to 4.6 hours, with Males having the highest average screen time on Smartphones and Females having the highest on Laptops.

# **WEEK 4: Advanced Visualizations**

# Day 1: Distribution plots with hue(Screen time by Gender)

```python
import seaborn as sns # Import the seaborn library for enhanced visualizations
import matplotlib.pyplot as plt # Import the matplotlib library for plotting

sns.displot(df, x='Avg_Daily_Screen_Time_hr', hue='Gender', kind='kde', fill=True, palette='coolwarm') # Create a displot (distribution plot) of 'Avg_Daily_Screen_Time_hr' with 'Gender' as hue, using a kernel density estimate, filling the areas under the curves, and using the 'coolwarm' color palette
plt.title("Distribution of Average Daily Screen Time by Gender") # Set the title of the plot
plt.show() # Display the plot
```

# Day 2: Stacked barplots(Device usage by location)

```python
import pandas as pd # Import the pandas library for data manipulation
import seaborn as sns # Import the seaborn library for enhanced visualizations
import matplotlib.pyplot as plt # Import the matplotlib library for plotting

# Calculate the counts of Primary_Device for each Urban_or_Rural category
device_location_counts = df.groupby('Urban_or_Rural')['Primary_Device'].value_counts().unstack(fill_value=0) # Group by 'Urban_or_Rural' and count 'Primary_Device', then unstack to get counts by location and device, filling missing values with 0

# Create a stacked barplot using pandas plot method
device_location_counts.plot(kind='bar', stacked=True, figsize=(8, 6), cmap='viridis') # Create a stacked bar plot using the calculated counts, with a specified figure size and colormap

plt.title('Device Usage by Location') # Set the title of the plot
plt.xlabel('Location (Urban or Rural)') # Set the label for the x-axis
plt.ylabel('Number of Users') # Set the label for the y-axis
plt.xticks(rotation=0) # Rotate x-axis labels for better readability
plt.legend(title='Primary Device') # Add a legend to the plot with the title 'Primary Device'
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show() # Display the plot
```

Insights: The stacked bar chart reveals that Smartphones are the most frequently used primary device for children in both Urban and Rural locations, but the overall volume of device usage is more than double in Urban areas (nearly 7,000 users) compared to Rural areas (under 3,000 users).

# Day 3:Swarmplots/Stripplots(Screen time spread)

```python
import seaborn as sns # Import the seaborn library for enhanced visualizations
import matplotlib.pyplot as plt # Import the matplotlib library for plotting

# Create a swarmplot to visualize the spread of daily screen time
plt.figure(figsize=(10, 6)) # Create a new figure for the plot with a specified size
sns.swarmplot(x=df['Avg_Daily_Screen_Time_hr']) # Create a swarmplot of the 'Avg_Daily_Screen_Time_hr' column
plt.title('Distribution of Average Daily Screen Time') # Set the title of the plot
plt.xlabel('Average Daily Screen Time (hr)') # Set the label for the x-axis
plt.show() # Display the plot
```

Insights: The Kernel Density Estimate (KDE) plot shows that the Average Daily Screen Time is unimodal with a strong peak around 5 hours.

# Day 4: Combine multiple plots into subplots

```python
import matplotlib.pyplot as plt # Import the matplotlib library for plotting
import seaborn as sns # Import the seaborn library for enhanced visualizations

# Create a figure with a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Histogram of Average Daily Screen Time
sns.histplot(data=df, x='Avg_Daily_Screen_Time_hr', kde=True, ax=axes[0, 0]) # Create a histogram of 'Avg_Daily_Screen_Time_hr' with a KDE curve on the first subplot
axes[0, 0].set_title('Distribution of Average Daily Screen Time') # Set the title for the first subplot
axes[0, 0].set_xlabel('Average Daily Screen Time (hr)') # Set the x-label for the first subplot

# Plot 2: Boxplot of Average Daily Screen Time by Gender
sns.boxplot(data=df, x='Gender', y='Avg_Daily_Screen_Time_hr', ax=axes[0, 1]) # Create a boxplot of 'Avg_Daily_Screen_Time_hr' by 'Gender' on the second subplot
axes[0, 1].set_title('Average Daily Screen Time by Gender') # Set the title for the second subplot
axes[0, 1].set_xlabel('Gender') # Set the x-label for the second subplot
axes[0, 1].set_ylabel('Average Daily Screen Time (hr)') # Set the y-label for the second subplot

# Plot 3: Countplot of Primary Device
sns.countplot(data=df, x='Primary_Device', ax=axes[1, 0], palette='viridis', hue='Primary_Device', legend=False) # Create a countplot of 'Primary_Device' on the third subplot
axes[1, 0].set_title('Count of Primary Device Usage') # Set the title for the third subplot
axes[1, 0].set_xlabel('Primary Device') # Set the x-label for the third subplot
axes[1, 0].set_ylabel('Count') # Set the y-label for the third subplot
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

# Plot 4: Scatterplot of Age vs Average Daily Screen Time
sns.scatterplot(data=df, x='Age', y='Avg_Daily_Screen_Time_hr', ax=axes[1, 1], alpha=0.6) # Create a scatterplot of 'Age' vs 'Avg_Daily_Screen_Time_hr' on the fourth subplot
axes[1, 1].set_title('Age vs Average Daily Screen Time') # Set the title for the fourth subplot
axes[1, 1].set_xlabel('Age') # Set the x-label for the fourth subplot
axes[1, 1].set_ylabel('Average Daily Screen Time (hr)') # Set the y-label for the fourth subplot


# Adjust layout to prevent titles/labels from overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

Insights:
* Most children have average daily screen time between 4 and 6 hours.

* The Smartphone is the most used primary device by children.

* Younger children (8-10) have the highest and most varied screen times, with some kids watching up to 14 hours.

* The amount of screen time is nearly the same for Male and Female children.

* Older children spend a smaller proportion of their screen time on educational activities.

* The total amount of screen time has very little connection to a child's age.

* Urban children are much more likely to use their screen time for educational activities than rural children.



**Week 4 insights:**
1. Day 1: Distribution Plots (Screen Time by Gender)

2. Day 2: Stacked Barplots (Device Usage by Location)
