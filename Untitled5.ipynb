```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Assuming you've already imported this and loaded your data into 'df'

# Load the data and define df (copied from cell VLin6TAaBdJ6)
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Indian_Kids_Screen_Time.csv")
df = data

## 1. Violin Plot: Screen Time by Gender 🧑‍🦱👩‍🦰

# Set the size of the plot
plt.figure(figsize=(8, 5))

# Create the Violin Plot
# 'x' is the categorical variable (Gender), 'y' is the continuous variable (Screen Time)
sns.violinplot(
    data=df,
    x='Gender',
    y='Avg_Daily_Screen_Time_hr', # Corrected column name
    palette='viridis',
    inner='quartile' # Displays the quartiles inside the violin
)

# Set the title and labels
plt.title('Violin Plot of Screen Time Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Screen Time (Hours/Day)')
plt.show()
```

```python
# Set the size of the plot (Wider figure for potentially more categories)
plt.figure(figsize=(12, 6))

# Create the Violin Plot
# 'x' is the categorical variable (Age), 'y' is the continuous variable (Screen Time)
sns.violinplot(
    data=df,
    x='Age',
    y='Avg_Daily_Screen_Time_hr', # Corrected column name
    palette='Spectral', # Use a different palette for contrast
    inner='box' # Displays a miniature box plot inside the violin
)

# Set the title and labels
plt.title('Violin Plot of Screen Time Hours by Age')
plt.xlabel('Age')
plt.ylabel('Screen Time (Hours/Day)')

# Rotate x-axis labels to prevent overlap if you have many unique ages
plt.xticks(rotation=45)
plt.tight_layout() # Adjusts plot parameters for a tight layout

plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
numeric_df = df.select_dtypes(include=['int64', 'float64'])
# Compute correlation matrix
corr = numeric_df.corr()
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap: Screen Time, Age, etc.", fontsize=14)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data (replace with your CSV file path)
df = pd.read_csv('Indian_Kids_Screen_Time.csv')

# Group by location and device, then count occurrences
device_counts = df.groupby(['Urban_or_Rural', 'Primary_Device']).size().unstack(fill_value=0)

# Plot stacked bar
device_counts.plot(kind='bar', stacked=True)
plt.xlabel('Location (Urban/Rural)')
plt.ylabel('Number of Kids')
plt.title('Device Usage by Location')
plt.legend(title='Device')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your CSV into a DataFrame
df = pd.read_csv("Indian_Kids_Screen_Time.csv")

# Basic swarmplot: Screen time by Gender
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.swarmplot(data=df, x="Gender", y="Avg_Daily_Screen_Time_hr")
plt.title("Screen Time Spread by Gender")
plt.show()
```
