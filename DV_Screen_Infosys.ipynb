```python
#Countplot for gender

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender')
plt.title('Count of Kids by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#countplot for device type
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='Primary_Device')
plt.title('Count of Kids by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate labels if they overlap
plt.show()
```

```python
#Countplot for location
plt.figure(figsize=(10,4))
sns.countplot(data=df, x='Urban_or_Rural')
plt.title('Count of Kids by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Violin plot for Screen Time by Gender
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x='Gender', y='Avg_Daily_Screen_Time_hr', palette="Set2")
plt.title('Daily Screen Time Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Avg Daily Screen Time (hrs)')
plt.show()
```

```python
#Violin plot for Screen Time by Age
plt.figure(figsize=(10,5))
sns.violinplot(data=df, x='Age', y='Avg_Daily_Screen_Time_hr', palette="muted")
plt.title('Daily Screen Time Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Avg Daily Screen Time (hrs)')
plt.show()
```

```python
# Boxplot: Screen Time by Age
plt.figure(figsize=(10, 6))
sns.boxplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df, palette="Set2")
plt.title("Boxplot of Screen Time by Age", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Average Daily Screen Time (hr)")
plt.show()
```

```python
# Boxplot: Screen Time by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x="Gender", y="Avg_Daily_Screen_Time_hr", data=df, palette="pastel")
plt.title("Boxplot of Screen Time by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Average Daily Screen Time (hr)")
plt.show()
```

```python
sns.set(style="whitegrid")

# Histogram for Age
plt.figure(figsize=(7, 5))
sns.histplot(df['Age'], bins=15, kde=False, color='skyblue')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Histogram for Average Daily Screen Time
plt.figure(figsize=(7, 5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], bins=15, kde=False, color='lightcoral')
plt.title("Histogram of Average Daily Screen Time (hr)")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Count")
plt.show()
```

```python
# KDE Plot for Age
plt.figure(figsize=(7, 5))
sns.kdeplot(df['Age'], fill=True, color='blue')
plt.title("KDE Plot of Age")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
# KDE Plot for Average Daily Screen Time
plt.figure(figsize=(7, 5))
sns.kdeplot(df['Avg_Daily_Screen_Time_hr'], fill=True, color='red')
plt.title("KDE Plot of Average Daily Screen Time (hr)")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Density")
plt.show()
```

```python
#Scatterplots (age vs screen time).
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df, hue='Gender', palette='coolwarm', s=70)
plt.title("Scatterplot: Age vs Average Daily Screen Time")
plt.xlabel("Age")
plt.ylabel("Average Daily Screen Time (hr)")
plt.legend(title="Gender")
plt.show()
```

```python
#Grouped barplots (weekday vs weekend by gender)
df['Day_Type'] = df['Exceeded_Recommended_Limit'].apply(
    lambda x: 'Weekend' if x else 'Weekday'
)

# Now plot grouped barplot
plt.figure(figsize=(8,6))
sns.barplot(
    x='Day_Type',
    y='Avg_Daily_Screen_Time_hr',
    hue='Gender',
    data=df,
    palette='Set2'
)

plt.title("Grouped Barplot: Weekday vs Weekend Screen Time by Gender")
plt.xlabel("Day Type")
plt.ylabel("Average Daily Screen Time (hr)")
plt.legend(title="Gender")
plt.show()
```

```python
#FacetGrid: Screen Time by Device × Gender
g = sns.FacetGrid(df, col="Gender", height=5, aspect=1)
g.map_dataframe(sns.barplot, x="Primary_Device", y="Avg_Daily_Screen_Time_hr", palette="Set2")
g.set_axis_labels("Primary Device", "Average Daily Screen Time (hr)")
g.set_titles(col_template="{col_name}")
g.fig.suptitle("Screen Time by Device x Gender", fontsize=14, y=1.05)
plt.show()
```

```python
#Pairplot: Relationships Between Numerical Variables
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numeric columns:", list(numeric_cols))
sns.pairplot(df[numeric_cols], diag_kind="kde", plot_kws={'alpha':0.6})
plt.suptitle("Pairplot: Relationships Between Numerical Variables", y=1.02)
plt.show()
```

```python
#Day 4: Heatmap (correlation between screen time, age, etc.).
numeric_df = df.select_dtypes(include=['int64', 'float64'])
# Compute correlation matrix
corr = numeric_df.corr()
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap: Screen Time, Age, etc.", fontsize=14)
plt.show()
```

```python
#Distribution Plot with Hue (Screen Time by Gender)
plt.figure(figsize=(8,6))
sns.histplot(
    data=df,
    x='Avg_Daily_Screen_Time_hr',
    hue='Gender',
    kde=True,                # Add smooth density curve
    bins=15,
    palette='Set2',
    alpha=0.6
)
plt.title("Distribution of Screen Time by Gender", fontsize=14)
plt.xlabel("Average Daily Screen Time (hours)")
plt.ylabel("Count")
plt.show()
```
