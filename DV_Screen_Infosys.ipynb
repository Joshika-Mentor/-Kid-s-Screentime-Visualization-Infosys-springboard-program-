Week 3 – Comparative Analysis

```python
#Grouped barplots (weekday vs weekend by gender)
df['Day_Type'] = df['Exceeded_Recommended_Limit'].apply(
    lambda x: 'Weekend' if x else 'Weekday'
)

# Now plot grouped barplot
plt.figure(figsize=(8,6))
sns.barplot(
    x='Day_Type',
    y='Avg_Daily_Screen_Time_hr',
    hue='Gender',
    data=df,
    palette='Set2'
)

plt.title("Grouped Barplot: Weekday vs Weekend Screen Time by Gender")
plt.xlabel("Day Type")
plt.ylabel("Average Daily Screen Time (hr)")
plt.legend(title="Gender")
plt.show()
```

```python
#FacetGrid: Screen Time by Device × Gender
g = sns.FacetGrid(df, col="Gender", height=5, aspect=1)
g.map_dataframe(sns.barplot, x="Primary_Device", y="Avg_Daily_Screen_Time_hr", palette="Set2")
g.set_axis_labels("Primary Device", "Average Daily Screen Time (hr)")
g.set_titles(col_template="{col_name}")
g.fig.suptitle("Screen Time by Device x Gender", fontsize=14, y=1.05)
plt.show()
```

```python
#Pairplot: Relationships Between Numerical Variables
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numeric columns:", list(numeric_cols))
sns.pairplot(df[numeric_cols], diag_kind="kde", plot_kws={'alpha':0.6})
plt.suptitle("Pairplot: Relationships Between Numerical Variables", y=1.02)
plt.show()
```

```python
#Day 4: Heatmap (correlation between screen time, age, etc.).
numeric_df = df.select_dtypes(include=['int64', 'float64'])
# Compute correlation matrix
corr = numeric_df.corr()
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap: Screen Time, Age, etc.", fontsize=14)
plt.show()
```

Week 4 – Advanced Visualizations

```python
#Distribution Plot with Hue (Screen Time by Gender)
plt.figure(figsize=(8,6))
sns.histplot(
    data=df,
    x='Avg_Daily_Screen_Time_hr',
    hue='Gender',
    kde=True,                # Add smooth density curve
    bins=15,
    palette='Set2',
    alpha=0.6
)
plt.title("Distribution of Screen Time by Gender", fontsize=14)
plt.xlabel("Average Daily Screen Time (hours)")
plt.ylabel("Count")
plt.show()
```

```python
#Day 2: Stacked barplots (device usage by location)
device_location = pd.crosstab(df['Urban_or_Rural'], df['Primary_Device'])
device_location.plot(     # Plot stacked bar chart
    kind='bar',
    stacked=True,
    figsize=(8,6),
    colormap='Set3'
)

plt.title("Device Usage by Location", fontsize=14)
plt.xlabel("Location Type")
plt.ylabel("Number of Users")
plt.legend(title="Primary Device", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Day 3: Swarmplots/stripplots (screen time spread)
plt.figure(figsize=(8,6))
sns.swarmplot(
    x='Gender',
    y='Avg_Daily_Screen_Time_hr',
    data=df,
    palette='Set2'
)
plt.title("Swarmplot: Screen Time by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Daily Screen Time (hrs)")
plt.show()
```

```python
#Stripplot: Screen Time by Primary Device
plt.figure(figsize=(8,6))
sns.stripplot(
    x='Primary_Device',
    y='Avg_Daily_Screen_Time_hr',
    data=df,
    jitter=True,
    alpha=0.7,
    palette='Set1'
)
plt.title("Stripplot: Screen Time by Primary Device")
plt.xlabel("Primary Device")
plt.ylabel("Average Daily Screen Time (hrs)")
plt.xticks(rotation=30)
plt.show()
```

```python
#Day 4: Combine multiple plots into subplots

fig, axes = plt.subplots(3, 2, figsize=(9, 8))
# 1. Histogram of Age
sns.histplot(df['Age'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title("Histogram: Age")

# 2. Histogram of Screen Time
sns.histplot(df['Avg_Daily_Screen_Time_hr'], kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title("Histogram: Screen Time")

# 3. Boxplot: Screen Time by Gender
sns.boxplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=df, ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title("Boxplot: Screen Time by Gender")

# 4. Countplot: Primary Device
sns.countplot(x='Primary_Device', data=df, ax=axes[1, 1], palette='Set3')
axes[1, 1].set_title("Countplot: Device Type")
axes[1, 1].tick_params(axis='x', rotation=20)

# 5. Scatterplot: Age vs Screen Time
sns.scatterplot(x='Age', y='Avg_Daily_Screen_Time_hr', hue='Gender', data=df, ax=axes[2, 0], palette='cool')
axes[2, 0].set_title("Scatterplot: Age vs Screen Time")

# 6. Violin plot: Screen Time by Location
sns.violinplot(x='Urban_or_Rural', y='Avg_Daily_Screen_Time_hr', data=df, ax=axes[2, 1], palette='pastel')
axes[2, 1].set_title("Violin Plot: Screen Time by Location")

plt.tight_layout()
plt.show()
```
