### Observations for Plots



**Day 1: Histograms & KDE plots for continuous variables (screen time, age).**

*   **Distribution of Average Daily Screen Time:** The distribution of average daily screen time appears to be roughly normally distributed, with a peak around 4-5 hours. There might be a slight skew to the right, indicating some children have significantly higher screen times.
*   **Distribution of Age:** The age distribution shows a relatively even spread across the age range in the dataset (8-18), with possibly slight peaks at certain ages.

**Day 2: Countplots for categorical variables (gender, device type, location).**

*   **Count Plot of Gender:** This plot shows the number of participants for each gender. Observe if the sample size is balanced between genders.
*   **Count Plot of Primary Device:** This plot reveals the most commonly used primary devices. Identify which devices are most prevalent among the children in the dataset.
*   **Count Plot of Urban or Rural:** This plot indicates the distribution of participants based on their location (urban or rural). Observe if there's a significant difference in the number of participants from urban versus rural areas.

**Day 3: Boxplots & violin plots (screen time by gender, by age group).**

*   **Box Plot of Average Daily Screen Time by Gender:** Compare the median screen time, the spread of data (interquartile range), and the presence of outliers for males and females.
*   **Box Plot of Average Daily Screen Time by Age:** Examine how the distribution of screen time varies across different age groups. Look for trends in median screen time, variability, and outliers as age increases.
*   **Violin Plot of Average Daily Screen Time by Gender:** This plot provides a richer view of the screen time distribution by gender than the box plot. Observe the density of the data at different screen time values for males and females.
*   **Violin Plot of Average Daily Screen Time by Age:** Similarly, this plot shows the distribution of screen time for each age. Look for multi-modal distributions or differences in spread across ages.

**Day 4: Scatterplots (age vs screen time).**

*   **Scatter Plot of Age vs. Average Daily Screen Time:** This plot helps visualize the relationship between age and screen time. Observe if there is a positive or negative correlation, if the relationship is linear, and if there are any clusters or outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Assuming you've already imported this and loaded your data into 'df'

# Load the data and define df (copied from cell VLin6TAaBdJ6)
data = pd.read_csv("/content/drive/MyDrive/Infosys_Springboard/Indian_Kids_Screen_Time.csv")
df = data

## 1. Violin Plot: Screen Time by Gender 🧑‍🦱👩‍🦰

# Set the size of the plot
plt.figure(figsize=(8, 5))

# Create the Violin Plot
# 'x' is the categorical variable (Gender), 'y' is the continuous variable (Screen Time)
sns.violinplot(
    data=df,
    x='Gender',
    y='Avg_Daily_Screen_Time_hr', # Corrected column name
    palette='viridis',
    inner='quartile' # Displays the quartiles inside the violin
)

# Set the title and labels
plt.title('Violin Plot of Screen Time Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Screen Time (Hours/Day)')

plt.show()

# ----------------------------------------------------

## 2. Violin Plot: Screen Time by Age 🎂

# Set the size of the plot (Wider figure for potentially more categories)
plt.figure(figsize=(12, 6))

# Create the Violin Plot
# 'x' is the categorical variable (Age), 'y' is the continuous variable (Screen Time)
sns.violinplot(
    data=df,
    x='Age',
    y='Avg_Daily_Screen_Time_hr', # Corrected column name
    palette='Spectral', # Use a different palette for contrast
    inner='box' # Displays a miniature box plot inside the violin
)

# Set the title and labels
plt.title('Violin Plot of Screen Time Hours by Age')
plt.xlabel('Age')
plt.ylabel('Screen Time (Hours/Day)')

# Rotate x-axis labels to prevent overlap if you have many unique ages
plt.xticks(rotation=45)
plt.tight_layout() # Adjusts plot parameters for a tight layout

plt.show()
```
