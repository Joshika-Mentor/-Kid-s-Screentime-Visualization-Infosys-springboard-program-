```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Assuming you've already imported this and loaded your data into 'df'

# Load the data and define df (copied from cell VLin6TAaBdJ6)
data = pd.read_csv("/content/drive/MyDrive/Infosys_Springboard/Indian_Kids_Screen_Time.csv")
df = data

## 1. Violin Plot: Screen Time by Gender 

# Set the size of the plot
plt.figure(figsize=(8, 5))

# Create the Violin Plot
# 'x' is the categorical variable (Gender), 'y' is the continuous variable (Screen Time)
sns.violinplot(
    data=df,
    x='Gender',
    y='Avg_Daily_Screen_Time_hr', # Corrected column name
    palette='viridis',
    inner='quartile' # Displays the quartiles inside the violin
)

# Set the title and labels
plt.title('Violin Plot of Screen Time Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Screen Time (Hours/Day)')

plt.show()

# ----------------------------------------------------

## 2. Violin Plot: Screen Time by Age 

# Set the size of the plot (Wider figure for potentially more categories)
plt.figure(figsize=(12, 6))

# Create the Violin Plot
# 'x' is the categorical variable (Age), 'y' is the continuous variable (Screen Time)
sns.violinplot(
    data=df,
    x='Age',
    y='Avg_Daily_Screen_Time_hr', # Corrected column name
    palette='Spectral', # Use a different palette for contrast
    inner='box' # Displays a miniature box plot inside the violin
)

# Set the title and labels
plt.title('Violin Plot of Screen Time Hours by Age')
plt.xlabel('Age')
plt.ylabel('Screen Time (Hours/Day)')

# Rotate x-axis labels to prevent overlap if you have many unique ages
plt.xticks(rotation=45)
plt.tight_layout() # Adjusts plot parameters for a tight layout

plt.show()
```

## Relationships involving health impacts

### Subtask:
Address missing values in the 'Health_Impacts' column and then explore the distribution of 'Health_Impacts' across other relevant variables.


**Reasoning**:
Inspect the unique values in the 'Health_Impacts' column to understand the different types of health impacts present and then fill the missing values with a placeholder.



```python
print(df['Health_Impacts'].unique())

# Fill missing values with 'No Health Impact Reported'
df['Health_Impacts'] = df['Health_Impacts'].fillna('No Health Impact Reported')

# Verify that missing values have been filled
print(df['Health_Impacts'].isnull().sum())
```

**Reasoning**:
Create count plots to show the distribution of 'Health_Impacts' across 'Gender' and 'Urban_or_Rural' after handling missing values.



```python
# Create a count plot for 'Health_Impacts' by 'Gender'
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Health_Impacts', hue='Gender', palette='viridis')
plt.title('Distribution of Health Impacts by Gender')
plt.xlabel('Health Impacts')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Create a count plot for 'Health_Impacts' by 'Urban_or_Rural'
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Health_Impacts', hue='Urban_or_Rural', palette='plasma')
plt.title('Distribution of Health Impacts by Urban or Rural Location')
plt.xlabel('Health Impacts')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Explore the relationship between 'Health_Impacts' and 'Primary_Device' using a count plot and then explore the relationship between 'Health_Impacts' and 'Age' by grouping age into ranges and using a count plot.



```python
# Create a count plot for 'Health_Impacts' by 'Primary_Device'
plt.figure(figsize=(16, 8))
sns.countplot(data=df, x='Health_Impacts', hue='Primary_Device', palette='tab10')
plt.title('Distribution of Health Impacts by Primary Device')
plt.xlabel('Health Impacts')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Group 'Age' into ranges
age_bins = [8, 11, 14, 18] # Define age ranges (e.g., 8-10, 11-13, 14-17)
age_labels = ['8-10', '11-13', '14-17']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Create a count plot for 'Health_Impacts' by 'Age_Group'
plt.figure(figsize=(16, 8))
sns.countplot(data=df, x='Health_Impacts', hue='Age_Group', palette='viridis')
plt.title('Distribution of Health Impacts by Age Group')
plt.xlabel('Health Impacts')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
