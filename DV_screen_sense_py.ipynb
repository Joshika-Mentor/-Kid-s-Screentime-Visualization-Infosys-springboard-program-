```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df="Indian_Kids_Screen_Time.csv"
df=pd.read_csv(df)
sns.violinplot(
    data=df,
    x="Gender",
    y="Avg_Daily_Screen_Time_hr",
    hue="Age",
    split=True,
    inner="quartile"
)
plt.title("Screen Time by Gender and Age")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,axes=plt.subplots(1,2,figsize=(16,6))
sns.histplot(data=df,x="Avg_Daily_Screen_Time_hr",kde=True,color='skyblue',ax=axes[0])
axes[0].set_title("Distribution of Average Daily Screen Time")
axes[0].set_xlabel("Average Daily Screen Time (hours)")
axes[0].set_ylabel("Frequency")

sns.histplot(data=df,x="Age",kde=True,color='lightcoral',ax=axes[1])
axes[1].set_title("Distribution of Age")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,axes=plt.subplots(1,2,figsize=(16,6))
sns.kdeplot(data=df,x='Avg_Daily_Screen_Time_hr',fill = True,color='skyblue',ax=axes[0])
axes[0].set_title("KDE Plot of Average Daily Screen Time")
axes[0].set_xlabel("Average Daily Screen Time (hours)")
axes[0].set_ylabel("Density")

sns.kdeplot(data=df,x='Age',fill=True,color='lightcoral',ax=axes[1])
axes[1].set_title("KDE Plot of Age")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = df


sns.set(style="whitegrid", palette="pastel")



data["Age_Group"] = pd.cut(
    data["Age"],
    bins=[5, 10, 13, 16, 19],
    labels=["6-10", "11-13", "14-16", "17-19"]
)

plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x="Age_Group", y="Avg_Daily_Screen_Time_hr", hue="Gender")

plt.title("Screen Time Distribution by Age Group and Gender", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Average Daily Screen Time (hours)")
plt.legend(title="Gender")
plt.show()
```

# **distribution plots with hue(screen_time by urban_or_rural)**



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = df


sns.set(style="white", palette="mako")

#  Histogram with KDE
plt.figure(figsize=(8, 5))
sns.histplot(
    data=data,
    x="Avg_Daily_Screen_Time_hr",
    hue="Urban_or_Rural",
    bins=20,
    kde=True,
    element="step"
)


plt.title("Distribution of Screen Time by Location (Urban vs Rural)", fontsize=14)
plt.xlabel("Average Daily Screen Time (hours)")
plt.ylabel("Count")
plt.legend(title="Location")
plt.show()


# KDE Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=data,
    x="Avg_Daily_Screen_Time_hr",
    hue="Urban_or_Rural",
    fill=True,
    common_norm=False,
    alpha=0.5
)

plt.title("Screen Time Density by Location (Urban vs Rural)", fontsize=14)
plt.xlabel("Average Daily Screen Time (hours)")
plt.ylabel("Density")
plt.legend(title="Location")
plt.show()


plt.figure(figsize=(8, 5))
sns.boxplot(
    data=data,
    x="Urban_or_Rural",
    y="Avg_Daily_Screen_Time_hr",
    palette="Set2"
)

plt.title("Screen Time Comparison by Location", fontsize=14)
plt.xlabel("Location")
plt.ylabel("Average Daily Screen Time (hours)")
plt.show()
```

# Stacked Barplots(Device usage by Location)

```python
import pandas as pd
import matplotlib.pyplot as plt


data = df


device_location = pd.crosstab(data['Urban_or_Rural'], data['Primary_Device'])


device_location.plot(
    kind='bar',
    stacked=True,
    figsize=(8, 5),
    colormap='viridis'
)

plt.title("Device Usage by Location")
plt.xlabel("Location")
plt.ylabel("Number of Users")
plt.legend(title="Primary Device", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Device usage by age**

```python
import pandas as pd
import matplotlib.pyplot as plt


data = df


device_age = pd.crosstab(data['Age'], data['Primary_Device'])


device_age.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap='tab20'
)


plt.title("Device Usage by Age")
plt.xlabel("Age")
plt.ylabel("Number of Users")
plt.legend(title="Primary Device", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
