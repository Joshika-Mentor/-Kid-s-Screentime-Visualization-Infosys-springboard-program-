### Facet Grid
#### We plot different plots i.e scatter, histogram, kde, etc. It gives different plots for our selected columns. Here we plot scatter plot for screentime for each device. It gives the scatter plot for each gender seperately. So that we can easily visualize and understand the data using this.

```python
g=sns.FacetGrid(df,col='Gender',hue="Age")
g.map(plt.scatter,'Primary_Device','Avg_Daily_Screen_Time_hr')
plt.show()
```

```python
g=sns.FacetGrid(df,col='Primary_Device',hue="Age")
g.map(plt.scatter,'Gender','Avg_Daily_Screen_Time_hr')
plt.show()
```

```python
g = sns.FacetGrid(df, col='Gender', row='Primary_Device', margin_titles=True, despine=False)
g.map(plt.hist, 'Avg_Daily_Screen_Time_hr', bins=20, color='skyblue', edgecolor='black')
g.fig.suptitle("Distribution of Average Daily Screen Time by Gender and Primary Device")
plt.tight_layout()
plt.show()
```

#### Above plots gives the distribution of screen time for each device by each gender seperately. By this we can draw the insights which device mostly used, screentime by each device and each gender screentime usage by each device. We see all of these using this single plot. Thats why facet grid is most useful as coming to plotting

### Heatmap
#### Heatmap defines the correlation between numerical columns in our data. It shows how much two columns are correlated i.e dependent on each other. The side color bar represents the color of lessly correlated to mostly correlated. The range varies from -1 to 1. 1 means positively correlated i.e one increases then another increases and -1 means negatively correlated viceversa to 1. 0 means not correlated. 

```python
plt.figure(figsize=(10,6))
numeric_columns=df.select_dtypes(include=['int64','float64'])
correlation_matrix=numeric_columns.corr()
sns.heatmap(correlation_matrix,cmap='coolwarm',annot=True)
plt.title("Heatmap:Correlation between numeric columns")
plt.tight_layout()
plt.show()
```
